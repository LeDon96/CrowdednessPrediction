# Data to File

## Wifi Sensors
Select Wifi sensors with objectnumber and coordinates, given that they are active

```python
#Save Wifi sensors in file
wifi_df = df.loc[(df["Soort"] == "WiFi sensor") & (df["Actief"] == "Ja")]
wifi_df = wifi_df[["Objectnummer", "LNG", "LAT"]]
wifi_df.to_csv("../Data/Modified/WifiSensors.csv")
```
